```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Definir las fases del proyecto y sus fechas de inicio y fin
fases = [
    {"fase": "Inicio y planificación", "inicio": "2026-01-01", "fin": "2026-03-31", "color": "#FF6347"},
    {"fase": "Desarrollo iterativo (8 sprints)", "inicio": "2026-04-01", "fin": "2026-12-31", "color": "#4682B4"},
    {"fase": "Pruebas piloto", "inicio": "2027-01-01", "fin": "2027-03-31", "color": "#32CD32"},
    {"fase": "Cierre y documentación", "inicio": "2027-04-01", "fin": "2027-06-30", "color": "#FFD700"}
]

# Convertir las fechas de inicio y fin a formato de fecha
for fase in fases:
    fase["inicio"] = pd.to_datetime(fase["inicio"])
    fase["fin"] = pd.to_datetime(fase["fin"])

# Crear el gráfico
fig, ax = plt.subplots(figsize=(12, 7))

# Añadir las barras de Gantt para cada fase
for i, fase in enumerate(fases):
    ax.barh(fase["fase"], (fase["fin"] - fase["inicio"]).days, left=fase["inicio"], height=0.6, 
            color=fase["color"], edgecolor="black", linewidth=1.2, zorder=10)

    # Añadir el texto dentro de la barra
    ax.text(fase["inicio"] + pd.Timedelta(days=(fase["fin"] - fase["inicio"]).days / 2), 
            i, fase["fase"], ha='center', va='center', fontsize=10, color="white", fontweight='bold')

# Ajustar las fechas en el eje x
ax.xaxis.set_major_locator(mdates.MonthLocator())  # Colocar los ticks por mes
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.xticks(rotation=45)

# Mejorar el formato de las fechas
plt.tick_params(axis="x", which="major", length=10, width=1.5, direction="in", grid_color="gray", grid_alpha=0.5)

# Añadir una cuadrícula sutil
ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Añadir etiquetas
#ax.set_xlabel('Tiempo', fontsize=12)
#ax.set_title('Cronograma Gantt - Plan de Proyecto', fontsize=14, fontweight='bold')

# Mejorar la visualización
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
